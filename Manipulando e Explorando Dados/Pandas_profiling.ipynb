A biblioteca Pandas Profiling é capaz de gerar um relatório em HTML com informações estatísticas e as principais características dos *Data Frame* utilizando apenas uma linha de código. 

Principais informações fornecidas pelo Pandas Profiling:

* Tamanho do *Data Set*;
* Quantidade de linhas duplicadas;
* Quantidade de linhas do *Data Frame*;
* Quantidade de linhas por colunas;
* Tipo de dado de cada coluna;
* Colunas que possuem ausência de valores;
* Distribuição das colunas numéricas.

Com o Pandas Profiling é possível gerar informações muito detalhadas em poucos minutos, aumentando a produtividade. 

Um problema dessa biblioteca é que para *Datasets* muito grandes a geração do relatório será muito demorada, para evitar esse problema pode ser feito uma análise de pedaços específicos do *Dataset*.

```python
DATA_PATH = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv"

#Importar bibliotecas necessárias
import pandas as pd
from pandas_profiling import ProfileReport

#Importar dataset
df = pd.read_csv(DATA_PATH)
```

```python
#Criar ProfileReport
report = ProfileReport(df)
```

```python
#Criar visualização iframe
report.to_notebook_iframe()
```

```python
#Criar relatório HTML
report.to_file('report.html')
```
