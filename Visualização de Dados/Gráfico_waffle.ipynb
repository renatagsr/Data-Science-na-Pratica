## Plotando um gráfico Waffle

A documentação da biblioteca pode ser encontrada [aqui](https://github.com/gyli/PyWaffle), nela há diversos exemplos de aplicação para seja mais fácil o aprendizado.

Os dados utilizados são fictícios, será simulado que foi realizado uma pesquisa no Brasil e que o resultado foi o seguinte: 33% falam bolacha, 48% falam biscoito e, uma minoria, 19% falem salgadinho. Com esses dados já é possível plotar o gŕafico.

Inicialmente, alguns poderiam pensar em plotar o gráfico de pizza, mas veja como esse gráfico fica muito melhor visualmente e muito mais fácil de entender do que um gráfico que precisa comparar áreas.

```python
#importar as biblioteas necessárias
import matplotlib.pyplot as plt
from pywaffle import Waffle

#Exemplo de Waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=4,
                columns=10,
                values={'Bolacha': 33,
                        'Biscoito': 48,
                        'Salgadinho': 19},
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1)}
)

fig.set_tight_layout(False)
plt.show()
```

## Customizando o Gráfico

Ao utilizar a biblioteca PyWaffle é possível customizar muito o gráfico, é possível alterar os ícones assim como seu tamanho, alterar as cores do gráfico, o número de colunas, colocar legenda, entre outras customizações. 

Agora será utilizado os dados proventintes da [base de dados do Centro de Prevenção e Investigação de Acidentes Aeronáuticos](http://www2.fab.mil.br/cenipa/index.php/estatisticas).


```python
# importar o pandas
import pandas as pd

# importar dados do CENIPA
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv")

# extrair value counts da variavel ocorrencia_classificacao
data = df.ocorrencia_classificacao.value_counts().to_dict()

# plotar gráfico de waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=5,
                columns=15,
                colors=("#232066", "#983D3D", "#DCB732"),
                values=data, icons='plane', icon_size=12,
                legend={'loc': 'upper left', 'bbox_to_anchor': (1,1)},
                icon_legend=True
                )
fig.set_tight_layout(False)
plt.show();
```

Como os dados utiilizados são os de acidentes aeronáuticos foram utilizados ícones de avião para enfatizar o que os dados representam, esse é mais um dos poderes de utilizar esse tipo de gráfico. 

```python
#Plotar homens e mulheres na sala de aula
fig = plt.figure(
    FigureClass=Waffle,
    rows=8,
    values={'Homens': 48, 'Mulheres': 58},
    colors=('#232066', '#983D3D'),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    icons='child', icon_size=12,
    icon_legend=True
)
fig.set_tight_layout(False)
plt.show()
```

Agora foi utilizado dados fictícios simulando o número de homens e mulheres que estão matrículados em um curso de inglês, visto que os dados são sobre pessoas foram utilizados ícones de pessoas e as cores escolhidas foram justamente para remeter que um é homem e outro é mulher. Essas customizações chamam muita a atenção do leitor e acaba facilitando o entendimento dele.
