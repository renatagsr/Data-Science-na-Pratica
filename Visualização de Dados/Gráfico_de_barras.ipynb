# Gráfico de Barras

Nele o comprimento de cada barra é proporcional ao valor que ela representa. É um tipo de gráfico comum para representar dados categóricos ou dados classificados em grupos. 

<center><img width=50% src='https://blog.sementenegocios.com.br/wp-content/uploads/2020/12/grafico-verticais-storytelling-1.jpg'></center>

[Fonte da Imagem](https://blog.sementenegocios.com.br/storytelling-com-dados/)

Esse gráfico é muito utilizado e é facilmente entendido pelas pessoas, porque os olhos dos seres humanos são muito bons para comparar comprimentos de objetos alinhados. Nossos olhos começam da base de cada barra e a examinam até o final, o comprimento é medido em relação à linha de base e às outras barras, dessa forma é um processo direto identificar a menor ou a maior barra. 

Há diversas variações comuns para esse gráfico:

* Barras verticas;
* Barras horizontais;
* Barras fixadas;
* Barras empilhadas;
* Barras 100% empilhadas;
* Waterfall.

Dicas para plotar um gráfico de barras:

* **Ajustar a distância entre as barras**: De forma geral, o espaço em branco entre as barras tem que ser menor do que a largura das barras, mas também o espaço não pode ser muito pequeno porque se não atrapalhará a leitura já que parecerá uma forma única.
* **Ordene as barras intencionalmente**: Se tiver uma ordem natural de ordenação, a utilize. Por exemplo: se estiver plotando gráficos com categorias de faixa etária, a ordenação deverá seguir a ordem crescente ou decrescente. Da mesma forma se a categoria for anos, tem que ser ordenado. A falta de ordenação fará com que o leitor tenha mais dificultade de interpretação.
* **Considere uma legenda**: Quando há mais de uma barra é necessário ter uma legenda para que o leitor entenda o que cada barra significa. Normalmente, os gráficos de barra não possuem muito espaço em branco então a rotulagem pode ser feita dentro das próprias barras.
* **Mova rótulos de dados diretos dentro das barras**: Se a rotulagem dos valores estiver sendo feito diretamente nas barras faça com que ela seja nas extremidades de cada barra, dessa forma os nossos olhos vão pensar que tem menos informação, facilitando a interpretação.
* **Use uma linha de base zero**: Toda a faixa de barras tem que estar presente, já que nos gráficos de barra os dados são codificados por comprimento.

## Plotar o Gráfico de Barras

Os dados desse *notebook* são dados públicos do Metrô de São Paulo. 

Os *datasets* podem ser encontrados no [Portal da Governança Corporativa e Transparência](https://transparencia.metrosp.com.br/).

O arquivo disponibilizado pelo *site* precisou ser estruturado e ele se encontra [neste link](https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/entrada_passageiro_linha1.csv).

```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

#Importar dados para dataframe
DATA_PATH = 'https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/entrada_passageiro_linha1.csv'
df = pd.read_csv(DATA_PATH)

#Ver as 5 primeiras entradas
df.head()
```

```python
#Ver as 5 últimas entradas
df.tail()
```

Percebe-se que ao final do *dataset* há uma linha TOTAL que não apresenta informações relevantes, pelo contratário, a sua existência pode atrapalhar a análise, principalmente a visual devido a discrepância do valor TOTAL para os demais valores. Portanto, será feita a retirada dessa linha.

```python
#Apagar linha TOTAL
df.drop(index=23, axis=0, inplace=True)
```

```python
#Plotar gráfico de barras com a entrada média de passageiros por estação
fig, ax = plt.subplots()

df.plot(x='Estação', y='Entradas', kind='bar', ax=ax, figsize=(16,8)) #Coloca o ax=ax para que seja plotado dentro do ax que foi criado anteriormente
ax.set_title('Entrada de Passageiros')
ax.set_xlabel('Estações')
ax.set_ylabel('Entradas')

plt.show()
```
