# Boxplot

Os *Boxplots* auxiliam a ver a distribuição dos dados, como eles estão agrupados, fornece informações estatísticas.

<center><image width=50% src='https://www.wellbeingatschool.org.nz/sites/default/files/W@S_boxplot-labels.png'></center>

[Fonte da Imagem](https://www.wellbeingatschool.org.nz/information-sheet/understanding-and-interpreting-box-plots)

Esse tipo de gráfico fornece 5 informações importantes:

* **Mediana**: Marca o ponto médio do conjunto de dados. É mostrada pela linha ao meio da caixa do gráfico, a que divide a caixa em duas partes. É o ponto que representa o 50%.
* **Primeiro Quartil**: O valor que representa a metade entre o menor NÚMERO e a MÉDIA do *dataset*. É o 25% dos dados.
* **Terceiro Quartil**: O valor que representa a metade entre o MAIOR número e a MÉDIA. É o 75%.
* **Interquartil (IQR)**: são os valores de 25% até 75% dos dados. É, basicamente, a caixa que é plotada. 
* **Maximum**: Q3(Terceiro Quartil) + 1,5 * IQR
* **Minimum**: Q1(Primeiro Quartil) - 1,5 * IQR.

[Fonte da imagem](https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51)

Boxplots são utilizados, de forma geral, para ver os padrões gerais de um grupo. Eles provém uma forma útil de visualizar a faixa de valores, se há presença de *outliers*, como os dados são distribuídos (de forma normal ou não). 

Alguns exemplos de identificação de padrões dos dados utilizando o *Boxplot*.

<center><image width=50% src='https://www.wellbeingatschool.org.nz/sites/default/files/W@S-different-boxplots.png'></center>

[Fonte da Imagem](https://www.wellbeingatschool.org.nz/information-sheet/understanding-and-interpreting-box-plots)


Pontos importantes sobre o Boxplot:

* **O boxplot é comparativamente curto**: no exemplo 2 (da imagem) sugere que de forma geral os alunos têm um alto nível de concordância uns com outros. Observe que os tamanho da caixa é bem menor do que as outras, isso mostra essa maior homogeneidade.
* **O boxplot é comparativamente alto**: Nos exemplos 1 e 3 mostra que a concordância dos alunos é pequena. 
* **As 4 seções do boxplot podem ser desiguais em tamanho**: No exemplo 1 fica vísivel esse ponto, cada uma das medidas tem um tamanho diferente, isso mostra uma distribuição de concordância diferentes entre o mesmo grupo. 
* **Mesma mediana, distribuição diferente**: Nos gráficos de 1 a 3 deles é bem semelhante umas as outras, estão próximas das médias também, mas percebe-se que a distribuição deles é totalmente diferentes.


## Plotar *Boxplot*

Para mostrar a forma de plotar um *boxplot* será utilizado um *dataset* contendo o número mensal de casos de dengue na cidade de Campinas-SP.

```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

#Importar dados para Dataframe
DATA_PATH = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv"
df = pd.read_csv(DATA_PATH)

#Ver 5 primeiras entradas
df.head()
```

```python
#Plotar o boxplot para as temperaturas
fig, ax = plt.subplots()

df.boxplot(['temperatura-media', 'temperatura-mininima', 'temperatura-maxima'], ax = ax)
plt.show()
```

Percebe-se que não há *outliers* nessas colunas selecionada para o conjunto de dados trabalhados.

```python
#Plotar o boxplot para casos confirmados
fig, ax = plt.subplots()

df.boxplot('casos-confirmados', ax = ax, vert=False)
plt.show()
```

Nesse caso, percebe-se que há muitos *outliers* visto que nem é possível ver o gráfico em si. Os valores discrepantes estão todos a direita do limite superior

```python
#Plotar o boxplot para casos confirmados, co corte arbitrário (<150)
fig, ax = plt.subplots()

df.loc[df['casos-confirmados'] < 80].boxplot('casos-confirmados', ax = ax, vert=False)
plt.show()
```

Fazendo a seleção fica vísivel a diferença que dá na visualização e na análise, agora é possível ver os valores do boxplot.
