O Histograma muitas vezes é confundido com o gráfico de barras, mas eles possuem uma grande diferença. ## Plotar o Histograma

Para exemplificar o gráfico de barras será utilizado um *dataset* contendo o número de casos mensais de dengue para a cidade de Campinas-SP. Esse conjunto de dados também contém informações a cerca da chuva e temperaturas. O período contemplado nesse *dataset* é de 1998 a 2015.

```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

#Importar dados para datafram
DATA_PATH = 'https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv'
df = pd.read_csv(DATA_PATH)
df.data = pd.to_datetime(df.data)
df.set_index('data', inplace=True)

#Ver as 5 primeiras entradas
df.head()
```

```python
#Plotar histogram de temperatura média
fig, ax = plt.subplots()

ax.hist('temperatura-media', data=df)
ax.set_title('Distribuição da Temperatura Média')
ax.set_ylabel('Quantidade de Ocorrências')
ax.set_xlabel('Temperatura média')

plt.show()
```

O histograma acima representa valores absolutos no eixo y.

```python
#Plotar histogram de temperatura média
fig, ax = plt.subplots()

ax.hist('temperatura-media', data=df, density=True)
ax.set_title('Distribuição da Temperatura Média')
ax.set_ylabel('Frequência de Ocorrências')
ax.set_xlabel('Temperatura média')

plt.show()
```

Para a plotagem desse histograma foi colocado o termo `density=True` para que os dados fossem plotados em relação a frequência relativa, em termos percentuais.

```python
#Plotar histogram de temperatura média
fig, ax = plt.subplots()

ax.hist('temperatura-media', data=df, density=True, bins = 40)
ax.set_title('Distribuição da Temperatura Média')
ax.set_ylabel('Frequência de Ocorrências')
ax.set_xlabel('Temperatura média')

plt.show()
```

A quantidade de barras que serão mostradas é definida por padrão, mas é possível alterar essa quantidade. No gráfico acima foi utilizado o `bins=40` para que a plotagem fosse feita com 40 barras. Pode fornecer qualquer número para o `bins`, mas é importante ficar atento que barras demais ou barras de menos podem fazer com que as informações visuais geradas percam o seu sentido.
